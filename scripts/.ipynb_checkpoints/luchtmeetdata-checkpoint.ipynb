```python
import pandas as pd
import glob, os 
import numpy as np
```

```python
# set filepath for airquality measure data
path = '/Users/miloubisseling/Documents/UvA/DataScience/BigData/Project/optimal-home-finder/data/data_air_quality'

# import all csv files and load into dataframe
df = pd.concat(map(lambda file: pd.read_csv(file, sep=';'), glob.glob(os.path.join(path, "*.csv"))))

# delete rows if value 2017 is missing
df = df.replace('-', np.nan)
df = df[pd.notnull(df['2017'])]

# set filepath for airquality station data and load csv into dataframe
stations = pd.read_csv('/Users/miloubisseling/Documents/UvA/DataScience/BigData/Project/optimal-home-finder/data/data_stations/meetstations_luchtkwaliteit.csv',sep=';')

# merge dataframes on station
pd.merge(df, stations, on='station')
```
