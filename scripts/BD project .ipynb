```python
def categorizing(column):
    values = []
    row = 0
    col_max = column.max()
    col_min = column.min()
    col_mean = column.mean()
    bound_low = col_min + ((col_mean - col_min)/2)
    bound_high = col_mean + ((col_max - col_mean)/2)
    for i in column:
        if column[row] < bound_low:
            value = 0
        elif column[row] > bound_high:
            value = 2
        elif column[row] > bound_low and column[row] < bound_high: 
            value = 1
        row += 1
        values.append(value)
#         values.append([value, bound_high, bound_low, col_max, col_min, col_mean])
    return values
```
